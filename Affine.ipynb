# **Encryption**

```python
import string

calcenc = lambda c, a, b: (a * c + b) % 26

def encryption(p, a, b):
    c = ""
    for i in p:
        if i.isalpha():
            if i in lower:
                c += lower[calcenc(lower.index(i), a, b)]
            else:
                c += upper[calcenc(upper.index(i), a, b)]
        else:
            c += i
    return c

if __name__ == "__main__":
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase

    # Get user input for plaintext and key values
    plaintext = input("Enter the plaintext: ")

    try:
        a = int(input("Enter the value of 'a' (an integer): "))
        b = int(input("Enter the value of 'b' (an integer): "))
    except ValueError:
        print("Error: 'a' and 'b' must be integers.")
        exit()

    # Encrypt the plaintext
    encrypted_text = encryption(plaintext, a, b)
    print("Encrypted text: " + encrypted_text)
```

# **Decryption**

```python
import string

calcenc = lambda c, a, b: (a * c + b) % 26
calcdec = lambda c, a, b: (keyinverse(a, 26) * (c - b)) % 26

def keyinverse(a, m):
    for i in range(1, m):
        if (((a % m) * (i % m)) % m == 1):
            return i
    return -1

def encryption(p, a, b):
    c = ""
    for i in p:
        if i.isalpha():
            if i in lower:
                c += lower[calcenc(lower.index(i), a, b)]
            else:
                c += upper[calcenc(upper.index(i), a, b)]
        else:
            c += i
    return c

def decryption(c, a, b):
    p = ""
    for i in c:
        if i.isalpha():
            if i in lower:
                p += lower[calcdec(lower.index(i), a, b)]
            else:
                p += upper[calcdec(upper.index(i), a, b)]
        else:
            p += i
    return p

if __name__ == "__main__":
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase

    # Get user input for encrypted text and key values
    encrypted_text = input("Enter the encrypted text: ")

    try:
        a = int(input("Enter the value of 'a' (an integer): "))
        b = int(input("Enter the value of 'b' (an integer): "))
    except ValueError:
        print("Error: 'a' and 'b' must be integers.")
        exit()

    # Decrypt the encrypted text
    decrypted_text = decryption(encrypted_text, a, b)
    print("Decrypted text: " + decrypted_text)
```

```python
import string

def keyinverse(a, m):
    for i in range(1, m):
        if (((a % m) * (i % m)) % m == 1):
            return i
    return -1

def decryption(ciphertext, a, b):
    p = ""
    for i in ciphertext:
        if i.isalpha():
            if i.islower():
                p += chr(((keyinverse(a, 26) * (ord(i) - ord('a') - b)) % 26) + ord('a'))
            elif i.isupper():
                p += chr(((keyinverse(a, 26) * (ord(i) - ord('A') - b)) % 26) + ord('A'))
        else:
            p += i
    return p

def brute_force_decrypt(ciphertext):
    lower = string.ascii_lowercase
    upper = string.ascii_uppercase
    results = []

    for a in range(26):
        for b in range(26):
            decrypted_text = decryption(ciphertext, a, b)
            results.append((a, b, decrypted_text))

    return results

if __name__ == "__main__":
    ciphertext = input("Enter the ciphertext: ")

    possible_decryptions = brute_force_decrypt(ciphertext)

    for a, b, text in possible_decryptions:
        print(f"Possible decryption (a={a}, b={b}): {text}")
```

# **Identifying The correct Sentences / Words**

```python
import string
import nltk
from language_tool_python import LanguageTool

nltk.download("punkt")

# Function to decrypt the text (similar to previous examples)
def decryption(ciphertext, a, b):
    p = ""
    for i in ciphertext:
        if i.isalpha():
            if i.islower():
                p += chr(((keyinverse(a, 26) * (ord(i) - ord('a') - b)) % 26) + ord('a'))
            elif i.isupper():
                p += chr(((keyinverse(a, 26) * (ord(i) - ord('A') - b)) % 26) + ord('A'))
        else:
            p += i
    return p

# Function to find grammatically correct sentences
def find_grammatically_correct_sentences(text, grammar_checker):
    sentences = nltk.sent_tokenize(text)  # Tokenize text into sentences
    valid_sentences = []

    for sentence in sentences:
        tokens = nltk.word_tokenize(sentence)
        if len(tokens) > 3:  # You can adjust this threshold as needed
            matches = grammar_checker.check(sentence)
            if len(matches) == 0:
                valid_sentences.append(sentence)

    return valid_sentences

if __name__ == "__main__":
    ciphertext = input("Enter the ciphertext: ")

    decrypted_texts = []  # Store all decrypted texts
    grammatically_correct_sentences = []  # Store grammatically correct sentences

    grammar_checker = LanguageTool('en-US')  # English grammar checker

    for a in range(26):
        for b in range(26):
            decrypted_text = decryption(ciphertext, a, b)
            decrypted_texts.append((a, b, decrypted_text))

    for a, b, text in decrypted_texts:
        valid_sentences = find_grammatically_correct_sentences(text, grammar_checker)
        if valid_sentences:
            print(f"Possible decryption (a={a}, b={b}):\n{text}")
            print("Grammatically correct sentences:")
            for sentence in valid_sentences:
                print(sentence)
            print("-----------")
```
